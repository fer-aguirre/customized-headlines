```python
from tqdm import tqdm
import customized_headlines.data.load as load
import customized_headlines.data.analyze as analyze
import customized_headlines.data.export as export
```

### Load data

```python
# Load data processed
news_sample = load.news_sample
news_content = load.news_content
```

### Read data

```python
# Read csv file as dataframe
df = load.read_csv(news_sample)
# Show dataframe
df.head(5)
```

```python
# Get unique values from a dataframe column
urls = analyze.get_col_val(df, 'archived_url')

# Create an empty list
texts = []
# Iterate through urls list
for i in tqdm(range(len(urls))):
    # Make a request
    request = analyze.make_request(urls[i])
    text = analyze.extract_text(request)
    texts.append(text)
```

```python
# Remove new lines from list
texts = [x.replace('\n', '') for x in texts]
```

```python
# Assign list to column in dataframe
df['text'] = texts


# Show dataframe
df.head(5)
```

### Exports

```python
# Save dataframe as csv
export.df_csv(df, news_content)
```
